```python
print("[*] Вивести перші рядки таблиці за допомогою методу head")
import pandas as pd
import re
tb = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',
                  match=re.compile(r'^Коефіцієнт народжуваності в регіонах України \(1950\—2019\)*$', re.I),
                  thousands='',
                  decimal=',')
df = tb[0]
df.head()
```

```python
print("[*] Визначте кількість рядків та стовпців у датафреймі (атрибут shape)")
df.shape
```

```python
print('[*] Замініть у таблиці значення "—" на значення NaN')
import numpy as np
df = df.replace("—", np.nan)
df
```

```python
print("[*] Визначте типи всіх стовпців за допомогою dataframe.dtypes")
df.dtypes
```

```python
print("[*] Замініть типи нечислових колонок на числові.")
ls = df.columns[df.isna().any()].tolist()
print(ls)
df = df.astype({c: "float" for c in ls})
df.dtypes
```

```python
print("[*] Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum), %")
print(round((pd.isnull(df).sum()/(df.shape[0]))*100, 2))
```

```python
print("[*] Видаліть з таблиці дані по всій країні, останній рядок таблиці.")
df.drop(df.tail(1).index, inplace=True)
df
```

```python
print("[*] Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)")
df = df.apply(lambda row: row.fillna(row[1:].mean()), axis=1)
df
```

```python
year = "2019"
print(f"[*] Отримайте список регіонів, де рівень народжуваності у {year} році був вищим за середній по Україні.")
mean = df[df.columns[1:]].mean().mean()
print("mean =", mean)
df.loc[df[year] > mean]
```

```python
year = "2014"
print(f"У якому регіоні була найвища народжуваність у {year} році?")
maxiny = df[year].max()
print(f"max_in_{year} = {maxiny}")
df.loc[df[year] == maxiny]
```

```python
# Відстань від регіонів України до росії (вибрана Казань як центр проекції для зменшення похибок)
dist=[1603.09, 1574, 1682.16, 1255.75, 1159.37, 1495.06, 1979.26, 1300.84, 1811.67, 1369.42, 1400.71, 1042.25, 1798.58, 1534.15, 1640.33,
      1197.29, 1619.86, 1094.36, 1720.82, 1076.88, 1531.86, 1647.50, 1346.29, 1781.33, 1261.98, 1369.42, 1603.09]
def norma(ls: list):
    vmin = min(ls)
    ls = [(v - vmin) for v in ls]
    vmax = max(ls)
    ls = [round(100 * (vmax - v) / vmax, 2) for v in ls]
    return ls

dist0 = norma(dist)
born2000 = norma(df["2000"].tolist())
born2012 = norma(df["2012"].tolist())
born2019 = norma(df["2019"].tolist())
dfdb = pd.DataFrame({"Регіон": df["Регіон"].tolist(),
                     "Distance": dist0,
                     "Born 2000": born2000,
                     "Born 2012": born2012,
                     "Born 2019": born2019})
dfdb.plot(title="Залежність народжуваності від приведеної віддаленості регіонів від росії, %", x="Регіон", rot=85, figsize=(13,5),
          subplots=True, grid=True, xticks = range(len(dist0)))
print("Чим далі від росії, тим вища народжуваність, і навпаки.")
```
