```python
print('[*] Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування".')
c0 = qr.columns[(pd.isnull(qr).sum() > 0)]
c0 = [c for c in c0 if not c.lower().startswith("яз")]
print(c0)
qr = qr.drop(c0, axis=1)
qr.head()
```

```python
print('[*] Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування".')
round(100 * pd.isnull(qr).sum() / qr.shape[0], 2)
```

```python
print("[*] Видаліть усі рядки у вихідній таблиці за допомогою методу dropna (де немає NaN?).")
qr = qr.dropna()
pd.isnull(qr).sum()
```

```python
print("[*] Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python.")
python_data = qr[qr["Язык.программирования"].str.contains('python', case=False)]
python_data.tail()
```

```python
print("[*] Визначте розмір таблиці python_data за допомогою -методу- атрибуту shape.")
python_data.shape
```

```python
python_data["Должность"].unique().tolist()
```

```python
print('[*] Використовуючи метод groupby, виконайте групування за стовпчиком "Посада".')
grs = qr.groupby(['Должность'])
grs.groups
```

```python
print('[*] Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних ' \
      'за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць".')
salary_range = grs.agg({'Зарплата.в.месяц': ['min', 'max']})
salary_range
```

```python
print('[*] Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. ' \
      'Використовуйте її для методу apply та створіть новий стовпчик "avg"')
def fill_avg_salary(position):
    return round(python_data[python_data['Должность'] == position]['Зарплата.в.месяц'].mean(), 2)
salary_range["Зарплата.в.месяц", "avg"] = [fill_avg_salary(k) for k in grs.groups.keys()]
salary_range
```

```python
print("[*] Збережіть отриману таблицю в CSV файл.")
salary_range.to_csv("salary_range.csv", index=False)
```

```python
print("[*] Додайте від 3 до 5 графіків функцій різного типу на ваш вибір.")
salary_range.plot.barh(title="Заробітна платня в місяць за посадами", ylabel="Посада", rot=15, width=0.75)
```

```python
dt = {"Посада": [k for k in grs.groups.keys()],
      "Кількість": [len(grs.groups[k]) for k in grs.groups.keys()]}
df = pd.DataFrame(dt)
df.plot.bar(title="Кількість робітників на посадах", x="Посада", y="Кількість", rot=75, width=0.8)
```

```python
gr_by_lang = qr.groupby("Язык.программирования")
gr_by_lang.groups
dt = {"Мова": [k for k in gr_by_lang.groups.keys()],
      "Популярність": [len(gr_by_lang.groups[k]) for k in gr_by_lang.groups.keys()]}
df_lang = pd.DataFrame(dt)
el = df_lang['Популярність'].tolist()
sumel = qr.shape[0]
explode = []
im = 0
for e in el:
    p = 1 - (e / sumel) - 0.6
    explode.append((p + 0.2 * im) / 1.5)
    im = (im + 1) % 5
print("Популярність мов програмування:")
df_lang.plot.pie(y="Популярність", x="Мова", figsize=(5,5), explode=explode,
                 legend=False, ylabel="", labels=df_lang["Мова"], startangle=35, labeldistance=1.2)
```
